```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from sklearn.metrics import pairwise
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np
```

```python
embedding = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
docs=[
    "This project utilizes a transformer-based causal language model to perform natural language generation tasks.",
    "The MiniMax M2.1 model is loaded using the AutoModelForCausalLM class due to its decoder-only architecture.",
    "Tokenization and text generation are handled through the Hugging Face Transformers library with remote code execution enabled.",
    "The system supports multiple use cases, including conversational AI, summarization through prompting, and text completion.",
    "Proper model–class alignment is essential to avoid configuration errors during model initialization."

]
```

```python
user_query = "Explain why AutoModelForCausalLM is required for decoder-only transformer models such as MiniMax M2.1."
```

```python
doc_embeddings =embedding.embed_documents(docs)
```

```python
user_query_embedding = embedding.embed_query(user_query)
```

```python
scores=cosine_similarity([user_query_embedding],doc_embeddings)[0]
```

```python
index,scores =sorted(list(enumerate(scores)),key =lambda x:x[1])[-1]
```

```python
print(user_query)
print(docs[index])
print('simiralityscore',scores)
```
